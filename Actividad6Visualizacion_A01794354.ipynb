```python
#6. Escala los datos, si consideras necesario
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
#GENERAMOS COPIA DEL DATASET
X = dataset.copy().drop('Y', axis= 1)
scaler.fit_transform(X)
```

```python
#7. Reduce las dimensiones con PCA, si consideras necesario
```

Aplicamos el PCA

```python
from sklearn.decomposition import PCA
modelo_PCA = PCA()
resultado_PCA = modelo_PCA.fit(X)
```

```python
resultado_PCA = modelo_PCA.fit(scaler.fit_transform(X))
```

Primero vamos a seleccionar todos los valores para poder ver cuales son los que explican más.

Importante tomar en cuenta que se tienen que estandarizar los datos para que las varianzas tengan relación entre si y no sea solamente los que tengan los valores más altos.

```python
exp_pca = pd.DataFrame({'Components':np.arange(1,24, dtype= int), 'Explicacion':resultado_PCA.explained_variance_ratio_, 'Var Acumulada': np.cumsum(resultado_PCA.explained_variance_ratio_)})
exp_pca.sort_values(by= 'Explicacion', ascending= False)
```

```python
#Determinar el numero optimo de componentes principales
```

De acuerdo a lo requerido dentro del ejercicio podriamos decir que para representar el 96% de la varianza de los datos podemos hacerlo con los 14 componentes principales.

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
pca2 = PCA(n_components= 14)
pca2.fit(X)
```

```python
pcCompDf = pd.DataFrame(pca2.components_.transpose(),
                        index = X.columns,
                        columns= [f'PC{i}' for i in range(1,15)])
pcCompDf
```

Para poder responder la pregunta de cuales fueron mas relevantes para cada componente es mas facil con un grafico.

```python

fig, ax = plt.subplots(figsize=(20, 10))
_ = sns.heatmap(abs(pcCompDf), cmap= 'YlGnBu', annot= True, fmt= '0.1f')
plt.title('Componentes Principales')
plt.yticks(rotation= 360)
```

Como podemos ver las variables más importantes, son las que justo nos salieron con menos correlación en la matríz de correlación.

Por lo que las variables más importantes podemos decir que son:

De x1 a x6  y de x14 a x23

GRAFICO DE VARIANZA ACUMULADA
De este de manera grafic apuedes concluir cual es acumulado de varianza de acuerdo al componente

```python
PC_components = np.arange(modelo_PCA.n_components_) + 1
#PC_components

_ = sns.set(style = 'whitegrid', 
            font_scale = 1.2
            )

fig, ax = plt.subplots(figsize=(10, 7))

_ = sns.barplot(x = PC_components, 
                y = modelo_PCA.explained_variance_ratio_, 
                color = 'b'
                )

_ = sns.lineplot(x = PC_components-1, 
                 y = np.cumsum(modelo_PCA.explained_variance_ratio_), 
                 color = 'black', 
                 linestyle = '-', 
                 linewidth = 2, 
                 marker = 'o', 
                 markersize = 8
                 )

plt.title('Grafico Acumulado')
plt.xlabel('N-th Principal Component')
plt.ylabel('Varianza explicada')
plt.ylim(0, 1)
plt.show()
```
